```python
dicomdir = os.path.abspath('dicom/')
dbgapdir = os.path.abspath('../../PNC/66680/NeurodevelopmentalGenomics/')
ncbidir = os.path.abspath('../../ncbi/dbGaP-14635/')
```

```python
subjids = np.load('output/train_subjids.npy')
```

```python
dbgap = glob.glob(os.path.join(dbgapdir, '*/*'))
```

```python
dicomtar = [os.path.basename(f) for f in glob.glob(os.path.join(dicomdir, '*_1.tar.gz'))]
```
